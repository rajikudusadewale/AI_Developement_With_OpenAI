# Whisper Model API: Transcription & Translation

##### Transcription: Speech to Text (STT) API

```python
import os
import openai
os.environ['OPENAI_API_KEY'] = '*******************'
openai.api_key = os.environ.get('OPENAI_API_KEY')
#creation of images


from openai import OpenAI
client = OpenAI()

audio_file= open(r'C:\Users\USER-PC\Desktop\Build_AI_APP\rr.mp3', "rb")
transcription = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file
)
print(transcription.text)
```

##### Translation using STT API

```python
# Translation
# It only support translation to English at this time

from openai import OpenAI
client = OpenAI()

audio_file= open(r"C:\Users\USER-PC\Desktop\Build_AI_APP\german.mp3", "rb")
translation = client.audio.translations.create(
  model="whisper-1", 
  file=audio_file
)
print(translation.text)
```

##### Text to Speech (TTS) API

```python

from openai import OpenAI
client = OpenAI()

input = """
Oyindamola mi owón,
Ìfẹ́ mi fún ẹ kò ní òpin. Ìwọ ni ayọ̀ àti ìmùóṣẹ́ mi. Mo fẹ́ràn rẹ bi ọmọ ìyá mi, àti pé ìwọ nìkan ni mo fẹ́ bá gbé ìgbésí ayé mi láéláé.
Ojú rẹ dàbí ìràwọ̀ tó ń tàn ní òfúrufú, ẹ̀rín rẹ sì mú ọkàn mi kún fún ìdùnnú. Ìrísí rẹ ní àárọ̀ kọ̀ọ̀kan jẹ́ ọ̀kan lára ohun tó dára jù lọ ní ayé mi.
Ní ọjọ́ tí a ṣègbéyàwó, mo ṣèlérí láti fẹ́ràn rẹ, láti būkún fún ọ, láti bọ̀wọ̀ fún ọ, àti láti wà níbẹ̀ fún ọ ní gbogbo ìgbà, bóyá nínú àìsàn tàbí ní ìlera, nínú òṣì tàbí ọrọ̀. Èmi yóò maa ṣe bẹ́ẹ̀ títí ayé ainípẹ̀kun.
Ìfẹ́ mi fún ọ jinlẹ̀ bíi Òkun, ó sì fẹ̀ bíi àwọn ọ̀run. Ọlọ́run ṣe àyànfẹ́ ìwọ àti èmi láti wà pọ̀, mo sì yóò maa dúpẹ́ fún Rẹ̀ ní ìdí eyi ní gbogbo ọjọ́ ayé mi.
Dájúdájú, aya mi owón, mo nífẹ̀ẹ́ rẹ láé láéláé.
Ọkọ rẹ tímọ́tímọ́
"""

response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input= input
)

response.stream_to_file("speech.mp3")
```
