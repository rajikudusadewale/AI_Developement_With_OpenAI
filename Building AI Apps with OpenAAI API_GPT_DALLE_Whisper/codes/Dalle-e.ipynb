## DALL-E : Generation of an Images

```python

import os
os.environ['OPENAI_API_KEY'] = '*****x'
openai.api_key = os.environ.get('OPENAI_API_KEY')
#creation of images


from openai import OpenAI
client = OpenAI()

response = client.images.generate(
    model="dall-e-3",
    prompt="A cute baby being fed",
    n=1,
    size="1024x1024",
    quality = 'standard'
)

image_url = response.data[0].url
print(image_url)
```

```python
response = client.images.generate(
  model="dall-e-3",
  prompt="A young Nigerian married man driving inside a benz car with cream interior design",
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url
print(image_url)
```

```python
response = client.images.generate(
  model="dall-e-3",
  prompt="A young Nigerian married woman in a white dress sitting opposite driver's side inside a benz car with cream interior design. She is the only one in the benz car, no driver with her",
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url
print(image_url)
```

```python
# To save the image as a png file

import requests
from urllib.parse import urlparse

# Parse the URL to get the file name
parsed_url = urlparse(image_url)
file_name = '9ja_woman'

# Save the image to the working directory
image_data = requests.get(image_url).content
with open(f"{file_name}.png", "wb") as f:
    f.write(image_data)

print(f"Image saved as {file_name}.png in the working directory.")
```

```python
# To display it on Jupyter notebook
from PIL import Image
from io import BytesIO

image = Image.open(BytesIO(image_data))
display(image)
```
